```python
review = {}

for col in df.columns:
    team = pd.DataFrame(df[col])
    for idx in team[team[col] == 0].index:
        try:
            review[col].append(df.iloc[int(idx)].Argonne)
        except KeyError:
            review[col] = []
            review[col].append(df.iloc[int(idx)].Argonne)
review    
```

```python
avg = {}

for k,v in review.items():
    avg[k] = sum(v)/len(v)
    
avg        
```
