```python
group_means = {}
group_names = df.columns
for group in group_names:
    m = df[df[group]==0].mean()
    group_means[group] = m
group_means = pd.DataFrame(group_means)
group_means.columns = ["AFRL rater","Argonne rater","Boeing rater","Delta Dental rater","Ford rater",
                      "Hope Village rater","Kellog's rater","Neogen rater","Old Nation rater","Qside rater",]
#Columns are how members rated other projects
group_means
```
