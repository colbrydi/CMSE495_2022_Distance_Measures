## Abstract
The data is a set of ratings, 0-9, of each capstone group. 0 means that the person who rated that group was part of the group, and the rest are rated by perceived similarity between projects (1 is most similar, 9 is least similar). Based on our analysis of the data, we find the most similar projects to our own based on our members' ratings of other teams.

## Data input


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance_matrix

dist_df = pd.read_csv("2022_Project_distance_Matrix.csv")
dist_df
```

## Data Modeling

This section that builds a model.  This doesn't have to be all that advanced. You should probably start with some simple statistical models such as an average distance matrix.

```python
dist_matrix = distance_matrix(dist_df.T, dist_df.T)

print(dist_matrix)

avg_qside = np.mean(dist_matrix[:-1, 9])
print(avg_qside)

for column_A in dist_df:
    col_df = dist_df[dist_df[column_A] == 0]
    print(column_A)
    for column_B in col_df:
        print (column_B, col_df[column_B].mean())
    print('\n')
```

## Data Visualization

This section make some graphs visualizing your results.  A distance matrix and/or network graph may be cool.  Think though the best way to show what you learned.

```python
import networkx as nx
import seaborn as sns

sns.heatmap(dist_matrix, cmap = "bone", xticklabels = dist_df.columns, yticklabels = dist_df.columns)
#plt.bar(range(10), dist_matrix[:, 9])


## Number of members on each team 

members = {}
for column in dist_df.columns:
    members[column] = sum(dist_df[column]==0)

fig = plt.figure(figsize = (25, 7))
ax = fig.add_subplot(1, 3, 1)
ax.bar(members.keys(), members.values(), 
        align='center', 
       color = 'lightblue', 
       edgecolor = 'black'
       )
ax.set_facecolor('aliceblue')
plt.xlabel('Team', size = 12)
plt.ylabel('Number of Responses', size = 12)
plt.title('Number of Response per Team', pad = 10, size = 15)
plt.xticks(rotation = 45)


## Average Rating per Team
average_rating = {}
for column in dist_df.columns:
    team_members = sum(dist_df[column]==0)
    average_rating[column] = sum(dist_df[column])/(len(dist_df[column])-team_members)

fig = plt.figure(figsize = (25, 7))
ax = fig.add_subplot(1, 3, 1)
ax.bar(average_rating.keys(), average_rating.values(), 
        align='center', 
       color = 'lightblue', 
       edgecolor = 'black'
       )
ax.set_facecolor('aliceblue')
plt.xlabel('Team', size = 12)
plt.ylabel('Average Distance', size = 12)
plt.title('Average Distance Rating by Team', pad = 10, size = 15)
plt.xticks(rotation = 45)


## Team We Rate as Closest to QSIDE (only 2 members voted?)

qside = dist_df[dist_df['Qside'] == 0]
for column in qside:
    print (column, qside[column].mean())
```
